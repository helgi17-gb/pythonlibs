```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv('../creditcard.csv')
```

```python
df.value_counts(normalize=True).head()
```

```python
df.info()
```

```python
pd.options.display.max_columns = 100
```

```python
df.head(n=10)
```

```python
X = df.drop(['Class'], axis=1)
```

```python
y = df[['Class']]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100, stratify = y)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
parameters = [{'n_estimators': [10, 15], 
               'max_features': np.arange(3, 5), 
               'max_depth': np.arange(4, 7)}]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
grid_search_cv = GridSearchCV(estimator = RandomForestClassifier(random_state=100),
                             param_grid = parameters, 
                             scoring = 'roc_auc', cv = 3)
```

```python
grid_search_cv.fit(X_train, y_train.values[:,0])
```

```python
grid_search_cv.best_params_
```

```python
grid_search_cv.predict_proba(X_train)
```

```python
y_pred_proba = grid_search_cv.predict_proba(X_train)[:,1]
```

```python
y_pred_proba.shape
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_train, y_pred_proba)
```

```python
y_pred_proba_test = grid_search_cv.predict_proba(X_test)[:,1]
```

```python
roc_auc_score(y_test, y_pred_proba_test)
```
